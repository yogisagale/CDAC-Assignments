### Q1. Write a password generator in Python. Be creative with how you generate passwords - strong passwords have a mix of lowercase letters, uppercase letters, numbers, and symbols. The passwords should be random, generating a new password every time the user asks for a new password. Include your code in a main method.
Extra:
* Ask the user how strong they want their password to be. For weak passwords, pick a word or two from a list.


```python
from random import choice as chc


def strong_password():
    '''Generaes a 8 digit password inclding upper case, lower case, digits, special chars'''
    u_caps = range(65, 91)
    l_caps = range(97, 113)
    spcl = range(33, 48)
    nums = range(48, 58)
    unfrmtd = [chc(u_caps), chc(spcl), chc(u_caps), chc(nums), chc(l_caps), chc(spcl), chc(nums), chc(l_caps)]
    frmtd = ''.join(chr(i) for i in unfrmtd)
    return frmtd
    

def password_generator(strong=True):
    if strong:
        return strong_password()
    else:
        return "apple"

pwd_type = input("Which type of password would you like (strong/weak) ? ").lower().strip()
strong_pwd = True if pwd_type == 'strong' else False

print("\nPassword Generated ==> ", password_generator(strong_pwd))
```

### Q3. Create a program that will play the “cows and bulls” game with the user. The game works like this:
### Randomly generate a 4-digit number. Ask the user to guess a 4-digit number. For every digit that the user guessed correctly in the correct place, they have a “cow”. For every digit the user guessed correctly in the wrong place is a “bull.” Every time the user makes a guess, tell them how many “cows” and “bulls” they have. Once the user guesses the correct number, the game is over. Keep track of the number of guesses the user makes throughout the game and tell the user at the end.
Say the number generated by the computer is 1038. An example interaction could look like this:
<pre><code>
Welcome to the Cows and Bulls Game! 
Enter a number: 
>>> 1234
2 cows, 0 bulls
>>> 1256
1 cow, 1 bull
...
</code></pre>
Until the user guesses the number.


```python
from random import randint


def cows_and_bulls():
    print('Welcome to the Cows and Bulls Game!')
    tries = 0
    bull = False
    while(not bull):
        tries += 1
        r_num = randint(1000, 9999)
        num = int(input("Guess the 4 digit number: "))
        if r_num == num:
            bull = True
        print("%d cows, %d bulls" % (tries-int(bull), int(bull)))

    print("Total attempts :", tries)
    
cows_and_bulls()
```

### Q4. Write a program that asks the user how many Fibonacci numbers to generate and then generates them. Take this opportunity to think about how you can use functions. Make sure to ask the user to enter the number of numbers in the sequence to generate.

```python
def fibonacci(num):
    fib_list = [1, 1]
    for i in range(num-2):
        fib_list.append(fib_list[i] + fib_list[i+1])
    return fib_list

num = int(input("Please enter the limiting number for Fibonacci Series : "))
print("Fibonacci Series upto %d numbers :" % num, fibonacci(num))
```

### Q5. Given two .txt files that have lists of numbers in them, find the numbers that are overlapping. One .txt file has a list of all prime numbers under 1000, and the other .txt file has a list of happy numbers up to 1000.


```python
#PRE-REQUISITES

# Generating first 1000 prime and happy numbers

from math import factorial


def wilson_s_formula(num):
    return True if factorial(num-1)%num == (num-1) else False


def sum_squares(num):
    num_sum = 0
    for i in str(num):
        num_sum += int(i)**2
    return int(num_sum)


def is_happy_number(num):
    while(num > 0):
        num = sum_squares(num)
        if num == 1 or num == 4:
            break 
            
    if num == 1:
        return True
    else:
        return False


def list_of_primes_and_happies(n):
    primes = []
    happy = []
    for i in range(1, n+1):
        if wilson_s_formula(i):
            primes.append(i)
        if is_happy_number(i):
            happy.append(i)
    return primes, happy      
```

```python
# Writing list to files

prime_list, happy_list =  list_of_primes_and_happies(1000)


with open('files/prime.txt', 'w') as prime, open('files/happy.txt', 'w') as happy:
    for num in prime_list:
        prime.write(str(num) + '\n')
    for num in happy_list:
        happy.write(str(num) + '\n')
```

```python
 # Now finding overlapping numbers by reading back from the file.
    
with open('files/prime.txt', 'r') as prime_file, open('files/happy.txt', 'r') as happy_file:
    prime_content = prime_file.readlines()
    happy_content = happy_file.readlines()
    prime_content = set(x.strip() for x in prime_content)
    happy_content = set(x.strip() for x in happy_content)

overlappers = prime_content.intersection(happy_content)
print("Total number of overlapping numbers are: ", len(overlappers))
print("\nNumbers that overlaps are the following :\n", ', '.join(overlappers))
```
