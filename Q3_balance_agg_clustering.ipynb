```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:\\Dataset\\balance.csv", header=None)
display(data.sample(5))
```

```python
data.info()
```

```python
print(data.iloc[:, 0].unique())
print(data.iloc[:, 1].unique())
print(data.iloc[:, 2].unique())
print(data.iloc[:, 3].unique())
print(data.iloc[:, 4].unique())
```

```python
from sklearn.preprocessing import LabelEncoder

leber = LabelEncoder()
data.iloc[:, 0] = leber.fit_transform(data.iloc[:, 0])
```

```python
display(data.head())
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
from sklearn.preprocessing import StandardScaler, Normalizer

X = StandardScaler().fit_transform(X)
X = Normalizer().fit_transform(X)
```

```python
from sklearn.decomposition import PCA 

pca = PCA(n_components = 2) 
X = pca.fit_transform(X)  
```

```python
from scipy.cluster.hierarchy import linkage

Z = linkage(X, method ='complete')
```

```python
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(12,12))
dendrogram(Z)
plt.title("Dendrogram")
plt.xlabel("Cluster Size")
plt.ylabel("Distance")
plt.show()
```

### Dendogram indicated n_cluster = 4 is a better choice

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=2)
plt.scatter(X[:,0], X[:,1], c = ac.fit_predict(X), cmap ='rainbow')
plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters=3)
plt.scatter(X[:,0], X[:,1], c = ac.fit_predict(X), cmap ='rainbow') 
plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters=4)
plt.scatter(X[:,0], X[:,1], c = ac.fit_predict(X), cmap ='rainbow') 
plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters=5)
plt.scatter(X[:,0], X[:,1], c = ac.fit_predict(X), cmap ='rainbow')
plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters=6)
plt.scatter(X[:,0], X[:,1], c = ac.fit_predict(X), cmap ='rainbow')
plt.show()
```

```python
from sklearn.metrics import silhouette_score 

k = [2, 3, 4, 5, 6]
sil_scr = [] 

for x in k:
    ac = AgglomerativeClustering(n_clusters = x) 
    sil_scr.append(silhouette_score(X, ac.fit_predict(X)))
```

```python
plt.bar(k, sil_scr) 
plt.xlabel('Number of clusters', fontsize = 20) 
plt.ylabel('Silhouette Scores', fontsize = 20) 
plt.show() 
```

### From dendogram and Silhouette Scores we can conclude n_cluster = 4 is a better choice
